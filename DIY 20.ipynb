```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# Q1. Problem Statement: Probability Distribution
# Load the exam.csv dataset into a DataFrame and perform the following tasks on
# the “math score,” ”reading score,” and “writing score” columns:
# 1. Plot the distributions using the seaborn library
# 2. Find the mean and standard deviation using the scipy library
# 3. Find skewness and kurtosis using the stats module from the scipy library
# 4. Find Standard Normal Distribution (z-score) using the stats module from the
# scipy library
# 5. Plot the Z distribution using the seaborn library
```

```python
data = pd.read_csv("exam.csv")
data
```

```python
data.info()
```

```python
data.describe()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
sns.distplot(data['math score'],hist = False, color='blue')
plt.title('Math Score Distribution')

plt.subplot(1, 3, 2)
sns.distplot(data['reading score'], hist = False,color='green')
plt.title('Reading Score Distribution')

plt.subplot(1, 3, 3)
sns.distplot(data['writing score'], hist = False, color='red')
plt.title('Writing Score Distribution')

plt.show()
```

```python
# 2. Find the mean and standard deviation using the scipy library

mean_math_score = data["math score"].mean()
mean_math_score
```

```python
mean_reading_score = data["reading score"].mean()
mean_reading_score
```

```python
mean_writing_score = data["writing score"].mean()
mean_writing_score
```

```python
# 3. Find skewness and kurtosis using the stats module from the scipy library

skewness_math = stats.skew(data["math score"])
kurtosis_math = stats.kurtosis(data['math score'])

skewness_reading = stats.skew(data['reading score'])
kurtosis_reading = stats.kurtosis(data['reading score'])

skewness_writing = stats.skew(data['writing score'])
kurtosis_writing = stats.kurtosis(data['writing score'])

print(f"Math Score: Skewness = {skewness_math}, Kurtosis = {kurtosis_math}")
print(f"Reading Score: Skewness = {skewness_reading}, Kurtosis = {kurtosis_reading}")
print(f"Writing Score: Skewness = {skewness_writing}, Kurtosis = {kurtosis_writing}")
```

```python
# 4. Find Standard Normal Distribution (z-score) using the stats module from the scipy library

data['z_math'] = stats.zscore(data['math score'])
data['z_reading'] = stats.zscore(data['reading score'])
data['z_writing'] = stats.zscore(data['writing score'])

# Display Z-scores
data[['z_math', 'z_reading', 'z_writing']]
```

```python
# 5. Plot the Z distribution using the seaborn library

sns.set(style="whitegrid")
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
sns.distplot(data['z_math'],hist = False, color='blue')
plt.title('Math Score Distribution')

plt.subplot(1, 3, 2)
sns.distplot(data['z_reading'], hist = False,color='green')
plt.title('Reading Score Distribution')

plt.subplot(1, 3, 3)
sns.distplot(data['z_writing'], hist = False, color='red')
plt.title('Writing Score Distribution')

plt.show()
```
